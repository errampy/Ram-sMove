# Merge / Join
**Join and Merge both are same  Join as SQL Terminology and Merge is Pandas Terminology**

### What is Join
*join is use to Joins the two  or more table's records is based on their relationship*

### Types of Join
    1. INNER JOIN
        Matching records from the both table is called inner join    
    2. OUTER JOIN
        Matching and non matching records from the both  table is called outer
    3. CROSS JOIN

#### Types of outer join
    2.1. Left Outer Join
        The records which are matching from both table and not matching from left table.
    2.2. Right Outer Join
        The records which are matching from both table and not matching fromo right table.
    2.3. Full Outer Join
        The records which are matching from both table and not not match records from the both tabel simple all records
        wheather its matches or not.

#### Example
***Examples are in below Cell***

```python
# lets create the DataSet of the list of employee and Department
emp_dataset = {
    'EMPNO':[101,102,103,104,105,106,107],
    'NAME' : ['A','B','C','D','E','F','G'],
    'SALARY': [1000,2000,3000,4000,5000,6000,7000],
    'DEPTNO' : [20,10,20,30,20,10,50]
}
dept_dataset = {
    'DEPTNO' : [10,20,30,40],
    'DNAME' : ['Z', 'Y', 'X', 'W'],
    'LOCATION' : ['Mumb', 'Chen', 'Delh', 'Punj']
}
```

```python
# lets import the library
import pandas as pd
```

```python
# Lets create the DataFrame of the dict of employee 
emp_df = pd.DataFrame(data=emp_dataset)
emp_df
```

```python
# Lets create the DataFrame of the dict of department
dept_df = pd.DataFrame(data=dept_dataset)
dept_df
```

```python
# Apply inner join
ij_data = pd.merge(left=emp_df, right=dept_df, how='inner', on='DEPTNO')
ij_data
```

```python
# Apply left outer join
loj = pd.merge(left=emp_df, right=dept_df, how='left', on='DEPTNO')
loj
```

```python
# Apply right outer join
roj = pd.merge(left=emp_df, right=dept_df, how='right', on='DEPTNO')
roj
```

```python
# Apply  full outer joing
roj = pd.merge(left=emp_df, right=dept_df, how='outer', on='DEPTNO')
roj
```

# If column names are not same then need to use extra parameter

left_on = 'name_of_column'
right_on = 'name_of_column'



```python
# lets create the DataSet of the list of employee and Department
emp_dataset = {
    'EMPNO':[101,102,103,104,105,106,107],
    'NAME' : ['A','B','C','D','E','F','G'],
    'SALARY': [1000,2000,3000,4000,5000,6000,7000],
    'DEPTNO' : [20,10,20,30,20,10,50]
}
dept_dataset = {
    'DEPARTMENT_NO' : [10,20,30,40],
    'DNAME' : ['Z', 'Y', 'X', 'W'],
    'LOCATION' : ['Mumb', 'Chen', 'Delh', 'Punj']
}
```

```python
emp_df = pd.DataFrame(data=emp_dataset)
emp_df
```

```python
dept_df = pd.DataFrame(data=dept_dataset)
dept_df
```

```python
# apply left outer join with different column name
pd.merge(left=emp_df, right=dept_df, how='left', left_on='DEPTNO', right_on='DEPARTMENT_NO')
```

```python
# apply right outer join with different column name
pd.merge(left=emp_df, right=dept_df, how='right', left_on='DEPTNO', right_on='DEPARTMENT_NO')
```

```python
# apply full outer join with different column name
pd.merge(left=emp_df, right=dept_df, how='outer', left_on='DEPTNO', right_on='DEPARTMENT_NO')
```

```python
pd.merge(left=emp_df, right=dept_df, left_on='DEPTNO', right_on='DEPARTMENT_NO', how='inner', validate='m:1', indicator=True)
```

```python
# apply indicator
# it will identify whether data is coming from left , right or both table

pd.merge(left=emp_df, right=dept_df, how='left', left_on='DEPTNO', right_on='DEPARTMENT_NO', indicator=True)
```
